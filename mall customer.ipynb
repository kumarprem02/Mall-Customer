# Explanatory Data Analysis:

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df,x='gender');
plt.title('Customer Gender Count',fontsize=20);
plt.xlabel('Gender',fontsize=20);
plt.ylabel('Count of Gender',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

**According to the visualization through the barplot the count of Male gender is less than the count of Female Gender.male customers are between 80-100 and female customers are more than 100**

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=df,x='gender',y='age');
plt.title('Comaprison of gender according to the age',fontsize=20);
plt.xlabel('Gender',fontsize=20);
plt.ylabel('Age',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

**The above plot is to illustrate age difference between the genders. looking at the visualization male age group is between 29-50 age and female age group is between 30-45 age. In male the age group between 37-50 age customers are more and in female the age group between 35-45 age customers  are more**

```python
plt.figure(figsize=(20,10))
sns.violinplot(data=df,x='gender',y='annual_income');
plt.title('Comparison of gender according to the Annual Income',fontsize=20);
plt.xlabel('Gender',fontsize=20);
plt.ylabel('Annual Income (K$)',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

**the above plot is to caompare the annual salary according to the age.**
**looking at the visualization annual income for male customers is higher than the female customers.** 
**median range of 75k dollors income are common from both the gender** 

```python
plt.figure(figsize=(20,10))
sns.violinplot(data=df,x='gender',y='spending_score');
plt.title('Comparison of gender according to the Spending Score',fontsize=20);
plt.xlabel('Gender',fontsize=20);
plt.ylabel('Spending Score',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
```

**the above plot is to illustrate the Comparison of gender according to the Spending score.looking at the visulaization both the genders have an equal height. but at a range between 40-60 the female customers are more than the male customers.**

```python
plt.figure(figsize=(40,10))
sns.boxplot(data=df,x='annual_income',y='spending_score',hue='gender');
plt.title('Comparison of Annual Income and Spending Score according to the age',fontsize=20);
plt.xlabel('Annual Income',fontsize=20);
plt.ylabel('Spending Score',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
plt.legend(fontsize=20);
```

```python
plt.figure(figsize=(40,10))
sns.boxplot(data=df,x='annual_income',y='age',hue='gender');
plt.title('Comparison of Annual Income and Age according to the gender',fontsize=20);
plt.xlabel('Age',fontsize=20);
plt.ylabel('Annual Income',fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
plt.legend(fontsize=20);
```

```python
plt.figure(figsize=(50,10))
sns.boxplot(data=df,x='spending_score',y='age',hue='gender');
plt.title('Comparison of Spending Score and Age according to the gender',fontsize=30);
plt.xlabel('Spending Score',fontsize=20);
plt.ylabel('Age',fontsize=20);
plt.xticks(fontsize=22);
plt.yticks(fontsize=25);
plt.legend(fontsize=25);
```
